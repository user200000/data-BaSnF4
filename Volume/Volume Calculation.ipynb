```python
import numpy as np
from pymatgen.io.vasp import Oszicar
import matplotlib.pyplot as plt
import ase.io
from figure_formatting import figure_formatting as ff

ff.set_formatting()
```

This book performs a birch murnaghan fit of the 

```python
enav=[]
vols=[]
for k in range(1,12):
    ennew=[]
    for l in range(4,5):
        PC=ase.io.read('SQS'+str(k)+'/r'+str(l)+'/POSCAR')
        en=Oszicar('SQS'+str(k)+'/r'+str(l)+'/Oszicar').all_energies
        for co in range(len(en)):
            ennew.append(en[co][-1])
        
    enav.append(np.mean(np.array(ennew)[:]))
    vols.append(PC.get_volume())
```

```python
from pymatgen.analysis.eos import BirchMurnaghan  
```

```python
bm=BirchMurnaghan(vols,enav)
```

```python
bm.fit()
```

```python
bm.plot()
```
